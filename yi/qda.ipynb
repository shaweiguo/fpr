```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import loadtxt
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.metrics import RocCurveDisplay
```

```python
trainA = loadtxt('data/classA_trn.dat')
trainB = loadtxt('data/classB_trn.dat')
testA = np.loadtxt('data/classA_test.dat')
testB = np.loadtxt('data/classB_test.dat')
```

```python
clf = QuadraticDiscriminantAnalysis()
```

```python
trainX = np.concatenate([trainA, trainB])
trainY = np.concatenate([np.ones((250, 1)), np.zeros((100, 1))])
```

```python
clf.fit(trainX, trainY)
```

```python
testYA = clf.predict(testA)
testYB = clf.predict(testB)
```

```python
X_test = np.concatenate([testA, testB])
Y_test = np.concatenate([np.ones((len(testA), 1)), np.zeros((len(testB), 1))])
clf_disp = RocCurveDisplay.from_estimator(clf, X_test, Y_test)
```
